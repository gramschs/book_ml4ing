```python
import numpy as np
import matplotlib.pylab as plt
import pandas as pd
import seaborn as sns
sns.set_theme()
```

```python
# import data
data_raw_01 = pd.read_csv('../../data/20220801_transfermarkt_bundesliga01.csv', skiprows=2, header=0, index_col=0)
data_raw_02 = pd.read_csv('../../data/20220801_transfermarkt_bundesliga02.csv', skiprows=2, header=0, index_col=0)
data_raw_03 = pd.read_csv('../../data/20220801_transfermarkt_bundesliga03.csv', skiprows=2, header=0, index_col=0)
data = pd.concat([data_raw_02, data_raw_03])

# encode categorical data
#data.replace('2. Bundesliga', 1, inplace=True)
#data.replace('3. Liga', 0, inplace=True)



```

```python
import sklearn

from sklearn.linear_model import Perceptron 
modell = Perceptron(eta0 = 1.0)
```

```python
X = data['Wert'].values.reshape(-1,1)
y = data['Ligazugehörigkeit'].values

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y)

modell.fit(X_train,y_train)
modell.score(X_test,y_test)
```
